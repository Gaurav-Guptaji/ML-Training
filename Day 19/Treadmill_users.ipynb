### Bivariate Analysis

1. boxplot          -  cat,num

2. violinplot       - cat,num
3. swarmplot        - cat,num (X)

4. barplot         - cat,num

5. jointplot        - num,num
6. scatterplot      - num,num (X)
7. regplot          - num,num

8. lineplot         - num,num
9. pointplot        - num,num

10. kdeplot          - num,num
11. rugplot         - num,num (X)


```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col}vs{ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.show()
```

```python
#sns.pairplot(data=data[num_cols])
pg=sns.PairGrid(data=data[num_cols],palette=choice(p_list))
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot)
```

```python
sns.heatmap(data=data[num_cols].corr(),annot=True,cmap=choice(p_list))
```

### Multivariate Analysis

```python
for col in cat_cols:
    if col == 'Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10, 3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1, 3, 1)
        sns.boxplot(
            data=data, x=col, y=ncol,
            hue='Product', palette=choice(p_list))
        plt.subplot(1, 3, 2)
        sns.violinplot(
            data=data, x=col, y=ncol,
            hue='Product', palette=choice(p_list))
        plt.subplot(1, 3, 3)
        sns.barplot(
            data=data, x=col, y=ncol,
            hue='Product', palette=choice(p_list))
        plt.show()
```
