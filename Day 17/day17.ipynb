```python
import numpy as np
```

### Types of graphs
1. plot -         line graph, between two parameters
2. bar -          bar graph of single parameter (optional categorical data)
3. hist -         histogram of a bunch of data i.e frequency of single parameter
4. scatter -      dot graph between two parameters
5. boxplot/ box - statistical overview in form of box for a single param
                  mean, outliners,quantiles(25%, 75%),IQR
                  (optional-catagorical data)
6. stem -         more like bar graph, line instead of bars
7. violinplot -   Violin shape statistical overview graph
                  spread data (more like histogram)
                  single parameter (optional cat_data)


```python
x=np.linspace(0,5,11)
x1=np.arange(1,11,3)
```

```python
x1
```

```python
y=x**3
```

```python
plt.plot(x,y)
plt.title('Plot Graph')
#plt.show() ---used when we write in script
```

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()
```

```python
plt.hist(x)
plt.title('Hisogram')
```

```python
plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x---->')
plt.ylabel('y---->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(x)
plt.title('Box Plot')
```

```python
plt.stem(x,y)
plt.title('Stem')
```

```python
plt.violinplot(x)
plt.title('Violinplot')
```

```python
y2=x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
#fig=plt.figure()---graph can be stored as pic on fig
plt.plot(x,y,lw=3,marker='o',markersize=12,markerfacecolor='yellow')
plt.plot(x,y2,marker='v',markersize=10)
plt.legend(['y=x**3','y=x**4'])
plt.grid()
plt.show()
```
